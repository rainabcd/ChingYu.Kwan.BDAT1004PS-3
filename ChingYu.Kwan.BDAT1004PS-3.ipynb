# Q1

```python
# Step 1. Import the necessary libraries
# Step 2. Import the dataset from this address.
# Step 3. Assign it to a variable called users
import pandas as pd

users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep="|")
print(users)
```

```python
# Step 4. Discover what is the mean age per occupation
users.groupby('occupation', as_index=False)['age'].mean()
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least 
male_users = users[users.gender=='M']
occ_male_count = male_users.groupby('occupation')['gender'].count()
occ_count = users.groupby(['occupation'])['gender'].count()
occ_male_r = ((occ_male_count/occ_count)*100).round(2)
occ_male_r = occ_male_r.sort_values(ascending=False)
print(occ_male_r)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
# reference: https://www.geeksforgeeks.org/pandas-groupby-one-column-and-get-mean-min-and-max-values/
users.groupby('occupation').agg({'age': ['min', 'max']})
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age
users.groupby(['occupation','gender'], as_index=True)['age'].mean()
```

```python
# Step 8. For each occupation present the percentage of women and men
occ_by_gender=users.groupby(['occupation','gender'])['gender'].count()
occ_count=users.groupby(['occupation'])['gender'].count()
gender_per_occ=((occ_by_gender/occ_count)*100).round(2)
gender_per_occ
```

```python
# Step 1. Import the necessary libraries
# Step 2. Import the dataset from the attached file wind.txt
# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.


import pandas as pd
import datetime

# reference: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html#:~:text=If%20True%20and%20parse_dates%20is,parsing%20speed%20by%205%2D10x.
data = pd.read_csv('wind.txt', sep="\s+", parse_dates = [[0,1,2]])
data
```

```python
# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
def fix_year(y):
    if y.year > 2000:
        year = y.year - 100
    else:
        year = y.year
    return datetime.date(year, y.month, y.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_year)
data
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

# reference: https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

data = data.set_index('Yr_Mo_Dy')

data
```

```python
# Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.
# reference: https://thispointer.com/python-pandas-count-number-of-nan-or-missing-values-in-dataframe-also-row-column-wise/

data.isnull().sum()
```

```python
# Step 7. Compute how many non-missing values there are in total.
data.notnull().sum()
```

```python
# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
# A single number for the entire dataset.
data.mean().mean()
```

```python
# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
# A different set of numbers for each location.
loc_stats = data.describe(percentiles=[])
loc_stats = loc_stats.drop(index=['count','50%'])
loc_stats
```

```python
loc_stats = data.apply(pd.Series.describe, axis=1)
loc_stats = loc_stats.drop(columns=['count','25%','50%','75%'])
loc_stats
```

```python
# Step 11. Find the average windspeed in January for each location.
# Treat January 1961 and January 1962 both as January.
jan_data = data[data.index.month == 1]
jan_data.mean()
```

```python
# Step 12. Downsample the record to a yearly frequency for each location.
data.resample('A').mean()
```

```python
# Step 13. Downsample the record to a monthly frequency for each location.
data.resample('M').mean()
```

```python
# Step 14. Downsample the record to a weekly frequency for each location.
data.resample('W').mean()
```

```python
#Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

loc_stats = data.apply(pd.Series.describe, axis=1)
loc_stats = loc_stats.drop(columns=['count','25%','50%','75%'])
weekly = loc_stats.resample('W').mean()
weekly.head(53)
```

```python
# Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
# Don't forget to label your axes!

import pandas as pd

marriage = pd.read_csv('us-marriages-divorces-1867-2014.csv', sep=",")
marriage
```

```python
m_years = marriage['Year'].values
m_per_cap = marriage['Marriages_per_1000'].values
d_per_cap = marriage['Divorces_per_1000'].values
```

```python
import matplotlib.pyplot as plt

plt.plot(m_years,m_per_cap,label='Marriage per capita')
plt.plot(m_years,d_per_cap,label='Divorce per capita')
plt.title('The number of marriages and divorces per capita in the U.S. between 1867 and 2014')
plt.xlabel('Year')
# plt.xlim([1867, 2014])
plt.legend()
plt.show()
```

```python
# Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
# Make sure that the pie chart is an even circle, labels the categories, and shows the percentage 
# breakdown of the categories.
roman = pd.read_csv('roman-emperor-reigns.csv', sep=",")
roman
```

```python
roman['Is_Assassinated']=roman['Cause_of_Death'].apply(lambda x: 'Y' if 'assassinated' in x.lower() else 'N')
roman
```

```python
Is_Assassinated_count = roman['Is_Assassinated'].value_counts()
Is_Assassinated_count
```

```python
plt.pie(Is_Assassinated_count, labels=['Other causes','Assassinated'],autopct='%1.1f%%')
plt.xlabel('Cause of Death')
plt.legend()

plt.show()
```
